### Resampling Time Series 

Resampling involves changing the frequency of your time series observations. Two types of resampling are: Upsampling: Where you increase the frequency of the samples, such as from minutes to seconds. Downsampling: Where you decrease the frequency of the samples, such as from days to months.

```python
#importing Libraries
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

Currently , we have data for each day.

```python
#daily
df.head()
```

```python
#Weekly
df.resample("W").last()
# here last method takes last day of every week . we can also use first() to take first day of every week.
```

```python
# if we want to have data of a particular day of week for example Wednesday
df.resample("W-wed").first()
# here we will get data for all wednesdays
```

```python
# we can also upscale from daily to hourly , but since in this dataset we dont have hourly data, you will see a NA in dataset
df.resample("H").first()
```

```python
# now if we want monthly data
df.resample("M").first()
```

```python
# but if we are analysing monthly data it will make more sense if we calculate mean of the value
df.resample("M").mean()
```

```python
#if we want first day of month as an index , 
df.resample("MS").mean()
# MS stands for MONTH-START
```

```python
#To get yearly data
df.resample("Y").mean()
```

```python
#if we want first date of year as an index , 
df.resample("YS").mean()
# MS stands for YEAR-START
```
