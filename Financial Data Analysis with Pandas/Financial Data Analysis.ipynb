```python
#importing Libraries
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#importing Data From Yahoo Finance
yf.download("AAPL", start = "2010-01-01", end = "2019-02-06")
```

```python
# we can also pass a list to extract data.
# apple , microsoft , ibm
list = ["AAPL" , "MSFT" , "IBM"]
df = yf.download(list ,start = "2010-01-01", end = "2019-02-06" )
```

```python
# we can also save this data into a csv file 
df.to_csv("stocks.csv")
```

```python
# now we can import the saved csv file
df = pd.read_csv("stocks.csv" , parse_dates = [0] , index_col = [0] , header = [0 , 1])
```

```python
df.head()
```

```python
#Convert a MultiIndex to an Index of Tuples containing the level values.
df.columns = df.columns.to_flat_index()
```

```python
#Convert flat index back to MultiIndex.
df.columns = pd.MultiIndex.from_tuples(df.columns)
```

```python
# we can swap the levels of headers. and sort the indexes to have all the data of one stock together.



df = df.swaplevel(axis = 1).sort_index(axis = 1)
```

```python
# for now we will be taking only Close column for visualisation.
close = df.loc[: , "Close"]
close.head()
```

```python
close.info()
```

```python
# plotting the close values of all three stocks
close.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

Here , we can see starting Close price of all three stocks are different so we have to normalize them to a base value

## Normalizing Time Series to a Base Value (100) 

```python
close.head()
```

```python
normalised_df = close.div(close.iloc[0]).mul(100)
close.iloc[0]
```

```python
normalised_df
```

```python
# plotting the normalised close values of all three stocks
normalised_df.plot(figsize = (15, 8 ), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

### Relative Difference between stock prices using shift method

Shift Method  , Shifts index by desired number of periods 

```python
close.head()
```

```python
close.shift(periods = 1)
```

```python
# Making a separate data frame for AAPL
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
#shifting 1 index
aapl["shifted"] = aapl.shift(periods = 1)
aapl
```

```python
#Now finding Relative difference between prices
aapl["diff"] = aapl.AAPL.sub(aapl.shifted)
aapl
```

```python
# we can also calculate percentage change
aapl["pct_change"] = aapl.AAPL.div(aapl.shifted).sub(1).mul(100)
aapl
```

```python
# python has methods diff() and pct_change from which we can directly calculate relative difference and percentage change
# diff method
close.diff()
```

```python
#pct_change
close.pct_change()
```
